```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('movies_dataset/tmdb_5000_movies.csv')
credits = pd.read_csv('movies_dataset/tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits, on='title')
```

```python
movies.head()
```

```python
# genres
# id
# keywords
# title
# overview
# popularity
# release_date
# cast
# crew

movies = movies[['movie_id', 'title', 'genres', 'overview', 'keywords', 'release_date', 'cast', 'crew', 'popularity']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
# '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# ['Action', 'Adventures', 'Fantasy', 'Science Fiction']
```

```python
import ast
def convert(list_of_obj):
    list_=[]
    for i in ast.literal_eval(list_of_obj):
        list_.append(i["name"])
    return list_
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies["keywords"] = movies["keywords"].apply(convert)
```

```python
import ast
def convert_3(list_of_obj):
    list_=[]
    cnt =  0
    for i in ast.literal_eval(list_of_obj):
        if cnt>2:
            break
        list_.append(i["name"])
        cnt+=1
    return list_
```

```python
movies["cast"] = movies["cast"].apply(convert_3)
```

```python
movies.head()
```

```python
def fetch_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job'] == "Director":
            l.append(i['name'])
            break
    return l
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies["genres"] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
movies["keywords"] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
movies["cast"] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movies["crew"] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id', 'title', 'tags']]
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: (" ".join(x)))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df.head()
```

```python
# stemming
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stemming(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stemming)
```

```python
# vectorization
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names()
```

```python
# Calculating distance 
#  - Euclidean distance
#  - Cosine distance
# Curse of dimensionality - Euclidean distance fails in more dimension
```

```python
# cosine similarity - 0 to 1 -- toward 1 than more similarity and towards 0 less similarity
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
def recommend_movies(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])
    top_five = movie_list[1:6]
    for i in top_five:
        print(new_df.iloc[i[0]].title)
```

```python
recommend_movies("Batman Begins")
```
