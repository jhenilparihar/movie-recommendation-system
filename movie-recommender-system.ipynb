```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('movies_dataset/tmdb_5000_movies.csv')
credits = pd.read_csv('movies_dataset/tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits, on='title')
```

```python
# genres
# id
# keywords
# title
# overview
# popularity
# release_date
# cast
# crew

movies = movies[['movie_id', 'title', 'genres', 'overview', 'keywords', 'release_date', 'cast', 'crew', 'popularity']]
```

```python
# '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# ['Action', 'Adventures', 'Fantasy', 'Science Fiction']
```

```python
import ast
def convert(list_of_obj):
    list_=[]
    for i in ast.literal_eval(list_of_obj):
        list_.append(i["name"])
    return list_
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies["keywords"] = movies["keywords"].apply(convert)
```

```python
import ast
def convert_3(list_of_obj):
    list_=[]
    cnt =  0
    for i in ast.literal_eval(list_of_obj):
        if cnt>2:
            break
        list_.append(i["name"])
        cnt+=1
    return list_
```

```python
movies["cast"] = movies["cast"].apply(convert_3)
```

```python
movies.head()
```

```python
def fetch_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job'] == "Director":
            l.append(i['name'])
            break
    return l
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies["genres"] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
movies["keywords"] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
movies["cast"] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movies["crew"] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id', 'title', 'tags']]
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: (" ".join(x)))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```
